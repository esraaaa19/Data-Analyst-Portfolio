```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/datasets/abhijitdahatonde/swiggy-restuarant-dataset/swiggy.csv')
df.head()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Basic Statistics
df.describe()
```

```python
print("Total Restaurants:", len(df))
```

```python
print("Average Rating:", round(df['Avg ratings'].mean(),2))
```

```python
df.sort_values(
    by='Total ratings',
    ascending=False
)[['Restaurant','Avg ratings']].head(10)
```

```python
df['Food type'].value_counts().head(10)
```

```python
df['Food type'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Cuisines")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['Avg ratings'], bins=20)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(df['Price'], bins=20)

plt.title("Restaurant Cost Distribution")
plt.xlabel("Cost")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['Price'], df['Avg ratings'])

plt.xlabel("Cost")
plt.ylabel("Rating")
plt.title("Cost vs Rating")
plt.show()
```
